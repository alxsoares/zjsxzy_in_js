# 读取word2vec模型

```python
model = gensim.models.Word2Vec.load("wallstreet_model")
```

```python
len(model.vocab) * 21
```

# 查看“重组”的相关词

```python
def isChinese(word):
    for w in word:
        if re.match('[ \u4e00 -\u9fa5]+',w) == None:
            continue
        else:
            return False
    return True
```

```python
key_word = u"江苏"
for w, v in model.most_similar(key_word, topn=15):
    if isChinese(w):
        print w, v
```

# “重组”的周热度变化

```python
with open("./word_count.json", 'r') as f:
    word_cnt = json.load(f)
```

```python
word_df = pd.DataFrame({"date": word_cnt[key_word].keys(), "count": word_cnt[key_word].values()})
```

```python
word_df.index = word_df["date"].map(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
word_df.sort_index(inplace=True)
```

```python
mean, std = word_df["count"].mean(), word_df["count"].std()
word_df["zscore"] = (word_df["count"] - mean) / std
```

```python
word_df["zscore"].plot()
```

```python
total_word_count = {}
for word in word_cnt:
    for day, value in word_cnt[word].iteritems():
        if not total_word_count.has_key(day):
            total_word_count[day] = 0
        total_word_count[day] += value * 1.0 / 10
```

```python
hot_val = {}
threshold = 0.1
for word in word_cnt:
    if isChinese(word) and word in model.vocab:
        dis = model.similarity(word, key_word)
        if dis > threshold:
            for day, value in word_cnt[word].iteritems():
                if not hot_val.has_key(day):
                    hot_val[day] = 0
                hot_val[day] += value * dis
for key, value in hot_val.iteritems():
    hot_val[key] = value / total_word_count[key]
```

```python
hot_df = pd.DataFrame({"date": hot_val.keys(), "value": hot_val.values()})
```

```python
hot_df.index = hot_df["date"].map(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
hot_df.sort_index(inplace=True)
```

```python
hot_df.drop(hot_df.index[-1], inplace=True)
```

```python
hot_df["value"].plot()
```

```python
mean, std = hot_df["value"].mean(), hot_df["value"].std()
```

```python
hot_df["zscore"] = (hot_df["value"] - mean) / std
```

```python
hot_df["zscore"].plot()
```

```python
df = pd.read_csv("./特斯拉.csv")
df.columns = ["date", "open", "high", "low", "close", "volume", "pvolume"]
df = df.dropna()
df.index = df["date"].map(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
```

```python
df["close"].plot()
```

```python
def word_index_plot(word, df, model, k=800):
    key_word = word
    for w, v in model.most_similar(key_word, topn=20):
        if isChinese(w):
            print w, v
    hot_val = {}
    threshold = 0.1
    for word in word_cnt:
        if isChinese(word) and word in model.vocab:
            dis = model.similarity(word, key_word)
            if dis > threshold:
                for day, value in word_cnt[word].iteritems():
                    if not hot_val.has_key(day):
                        hot_val[day] = 0
                    hot_val[day] += value * dis
    for key, value in hot_val.iteritems():
        hot_val[key] = value / total_word_count[key]
    
    hot_df = pd.DataFrame({"date": hot_val.keys(), "value": hot_val.values()})
    hot_df.index = hot_df["date"].map(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
    hot_df.sort_index(inplace=True)
    hot_df.drop(hot_df.index[-1], inplace=True)
    
    start_index = max(df.index[0], hot_df.index[0])
    df = df[df.index >= start_index]
    hot_df = hot_df[hot_df.index >= start_index]
    
    mean, std = hot_df["value"].mean(), hot_df["value"].std()
    hot_df["zscore"] = (hot_df["value"] - mean) / std
    df["close"].plot()
    (hot_df["zscore"] * k).plot()
    plt.legend(["index", "news"], loc='upper left')
```

```python
word_index_plot(u"特斯拉", df, model)
```
