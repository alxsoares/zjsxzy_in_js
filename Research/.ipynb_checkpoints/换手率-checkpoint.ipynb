```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
def wind2df(raw_data):
    dic = {}
    for data, field in zip(raw_data.Data, raw_data.Fields):
        dic[str(field.lower())] = data
    return pd.DataFrame(dic, index=raw_data.Times)
```

```python
data = w.wsd('881001.WI', 'close', '2015-01-01', '2017-05-17')
```

```python
price = wind2df(data)
```

```python
price['ret'] = price['close'].pct_change(20).shift(-20)
```

# 换手率的变化

```python
STOCK_DIR = 'D:Data/stocks'
COST_STOCK_DIR = 'D:Data/avg_cost/by stock'
```

```python
files = [f for f in os.listdir(STOCK_DIR)]
```

```python
dic = {}
for f in files:
    fname = '%s/%s'%(STOCK_DIR, f)
    df = pd.read_excel(fname, index_col=0)
    dic[f[:-5]] = df[['amt', 'mkt_freeshares']]
pnl = pd.Panel(dic)
```

```python
pnl.ix[:, :, 'turnover'] = pnl.minor_xs('amt') / pnl.minor_xs('mkt_freeshares')
```

```python
turnover = pnl.minor_xs('turnover').mean(axis=1)
```

```python
plt.figure(figsize=(16, 9))
turnover.plot()
```

# 换手天数

```python
files = [f for f in os.listdir(COST_STOCK_DIR)]
```

```python
dic = {}
for f in files:
    fname = '%s/%s'%(COST_STOCK_DIR, f)
    df = pd.read_excel(fname, index_col=0)
    dic[f[:-5]] = df[['turnover days']]
pnl = pd.Panel(dic)
```

```python
df = pnl.minor_xs('turnover days')
```

```python
df.tail()
```

```python
days = df.mean(axis=1)
```

```python
plt.figure(figsize=(16, 9))
days[days.index >= '2016-06-01'].plot()
```

```python
price['turnover'] = days
price['ret'] = price['close'].pct_change(60).shift(-60)
```

```python
sns.lmplot('turnover', 'ret', data=price)
```
