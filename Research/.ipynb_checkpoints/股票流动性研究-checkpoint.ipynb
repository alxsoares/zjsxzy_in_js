```python
from WindPy import w
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import preprocessing
import seaborn as sns
%matplotlib inline
```

```python
w.start()
```

```python
def wind2df(raw_data):
    dic = {}
    for data, field in zip(raw_data.Data, raw_data.Fields):
        dic[str(field.lower())] = data
    return pd.DataFrame(dic, index=raw_data.Times)
```

```python
BY_STOCK = "D:/data/"
code = '600000.SH'
start_date = '2015-01-01'
end_date = '2017-03-27'
```

# 股票流动性指标
* 成交额
* 成交笔数
* 换手率
* 成交量
* 主动买入额
* 主动卖出额

```python
fields = "amt,dealnum,free_turn,volume,mfd_buyamt_a,mfd_sellamt_a"
```

```python
data = w.wsd(code, fields, start_date, end_date, "traderType=1")
```

```python
df = wind2df(data)
```

```python
df.head()
```

# 近一个月成交额

```python
df['20-day amt'] = df['amt'].rolling(window=20).sum()
```

# 近一个月换手率

```python
df['20-day free_turn'] = df['free_turn'].rolling(window=20).sum()
```

# 流动性指标
* 近一个月成交额
* 近一个月换手率
* 成交量
* 成交笔数
* 主动买入额
* 主动卖出额

```python
indicators = [ '20-day amt', '20-day free_turn', 'volume', 'dealnum', 'mfd_buyamt_a', 'mfd_sellamt_a']
```

```python
X = df[indicators].dropna()
```

```python
X.tail()
```

# 标准化

```python
scaler = preprocessing.StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
X.shape
```

# PCA

```python
pca = PCA()
```

```python
pca.fit(X_scaled)
```

### 主成分贡献率

```python
pca.explained_variance_ratio_
```

### Eigen value

```python
centered_matrix = X - X.mean(axis=0)
cov = np.dot(centered_matrix.T, centered_matrix)
eigvals, eigvecs = np.linalg.eig(cov)
```

```python
eigvals
```

### eigen value从大到小排序的序号

```python
index = np.argsort(eigvals)[::-1]
```

```python
index
```

# 计算流动性指标
$$F = \sum_{k=1}^{m} \frac{\lambda_k}{\sum_{i=1}^{m}\lambda_k}\times F_k$$

```python
m = 2
```

```python
lambdas = eigvals[index[:m]]
```

```python
lambdas = lambdas / lambdas.sum()
```

```python
F = X_scaled[:, index[:m]] * lambdas
```

```python
F_score = F.sum(axis=1)
```

```python
Fdf = pd.DataFrame({'F': F_score}, index=X.index)
```

```python
plt.figure(figsize=(16, 9))
Fdf['F'].plot()
```
