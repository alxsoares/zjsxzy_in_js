```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
def wind2df(raw_data):
    dic = {}
    for data, field in zip(raw_data.Data, raw_data.Fields):
        dic[str(field.lower())] = data
    return pd.DataFrame(dic, index=raw_data.Times)
```

```python
data = w.wsd('881001.WI', 'close', '2015-01-01', '2017-05-09')
```

```python
df = wind2df(data)
```

```python
df['month ret'] = df['close'].pct_change(20).shift(-20)
```

```python
pnl = pd.read_pickle('C:/Users/jgtzsx01/Documents/workspace/zjsxzy_in_js/website/everyday-update/data/price.pkl')
```

```python
pnl.ix[:, :, 'month'] = pnl.minor_xs('close').rolling(window=20).mean()
```

```python
price = pnl.minor_xs('close') > pnl.minor_xs('month')
```

```python
array = price.sum(axis=1)
```

```python
df['month'] = array.values
```

```python
df.head()
```

```python
plt.figure(figsize=(16, 9))
df[df['month'] > 0]['month'].plot()
```

```python
sns.lmplot(x='month', y='month ret', data=df[df['month'] > 0])
```

```python
pnl.ix[:, :, 'year'] = pnl.minor_xs('close').rolling(window=242).mean()
```

```python
price = pnl.minor_xs('close') > pnl.minor_xs('year')
```

```python
array = price.sum(axis=1)
```

```python
df['year'] = array.values / (~pnl.minor_xs('close').isnull()).sum(axis=1)
```

```python
df.tail()
```

```python
plt.figure(figsize=(16, 9))
df[df.index >= '2016-03-01']['year'].plot()
```

```python
sns.lmplot(x='year', y='month ret', data=df[df.index >= '2016-03-01'])
```

```python
pnl.ix[:, :, 'quarter'] = pnl.minor_xs('close').rolling(window=60).mean()
```

```python
price = pnl.minor_xs('close') > pnl.minor_xs('quarter')
```

```python
array = price.sum(axis=1)
```

```python
df['quarter'] = array.values / (~pnl.minor_xs('close').isnull()).sum(axis=1)
```

```python
df.tail()
```

```python
plt.figure(figsize=(16, 9))
df[df.index >= '2016-03-01']['quarter'].plot()
```

```python
sns.lmplot(x='quarter', y='month ret', data=df[df.index >= '2016-03-01'])
```

```python
pnl.ix[:, :, 'ret'] = pnl.minor_xs('close').pct_change()
```

```python
array = (pnl.minor_xs('ret') > 0.09).sum(axis=1) / (~pnl.minor_xs('close').isnull()).sum(axis=1)
```

```python
plt.figure(figsize=(16, 9))
array.plot()
```

```python
df['up'] = array
```

```python
sns.lmplot(x='up', y='month ret', data=df[df.index >= '2016-03-01'])
```
