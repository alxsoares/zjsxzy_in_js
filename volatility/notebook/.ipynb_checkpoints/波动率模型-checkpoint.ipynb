### 1. 统计PCA行业集中度

```python
files = ["./industry_day/" + f for f in os.listdir("./industry_day/") if f.endswith(".csv")]
```

```python
def get_return(fp, date="2016-09-09", k=7):
    """
    获取fp文件中至date为止之前k天之前的收益率
    收益率计算公式为r_t=log(p_{t+1}) - log(p_t)
    """
    date = datetime.datetime.strptime(date, "%Y-%m-%d")
    df = pd.read_csv(fp)
    df.columns = ["date", "open", "high", "low", "close", "volume", "pvolume"]
    df = df.drop(df.index[-2:])
    df.index = df["date"].map(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
    df = df[df.index > datetime.datetime(2006, 1, 4)]
    df = df[(df.index >= date - datetime.timedelta(k)) & (df.index < date)]
    ret = np.diff(np.log(df["close"])) # log收益率: log(p_{t+1}) - log(p_t)
    return ret
```

```python
ret = get_return(files[0], date="2016-02-01", k=30)
```

```python
ret
```

```python
def get_all_matrix(files, date="2016-09-09", k=7):
    """
    获取files集合中至date位置之前k天的n个行业的收益率矩阵
    """
    mat = []
    for f in files:
        ret = get_return(f, date, k)
        mat.append(ret)
    return np.array(mat)
```

```python
mat = get_all_matrix(files, date="2016-02-01", k=30)
```

```python
mat.shape
```

```python
def get_eig_values(mat):
    """
    求矩阵mat的协方差矩阵，得到特征值向量
    """
    cov_mat = np.cov(mat)
    evals = np.linalg.eigvalsh(cov_mat)
    evals.sort()
    return evals[::-1]
```

```python
mat = get_all_matrix(files, date="2008-02-01", k=30)
```

```python
evals = get_eig_values(mat)
print np.sum(evals[:1]) / evals.sum()
print evals.max() / evals.sum()
```

```python
def get_volatility(df, date="2016-09-01", k=30):
    """
    获取date当日的波动率：计算之前k天的收益率的方差
    """
    date = datetime.datetime.strptime(date, "%Y-%m-%d")
    df = df[(df.index >= date - datetime.timedelta(k)) & (df.index < date)]
    ret = np.diff(np.log(df["close"])) # log收益率: log(p_{t+1}) - log(p_t)
    return np.std(ret)
```

# 周波动率建模

```python
current_date = datetime.datetime(2006, 12, 1)
df = csi[csi.index > current_date]
```

```python
volatility_week = {}
volatility_month = {}
volatility_year = {}
for i in df.index:
    current_week = current_date.isocalendar()[1]
    week = i.isocalendar()[1]
    week_day = i.isocalendar()[2]
    if week == current_week and week_day == 5: # 以周五的波动率代表一周
        volatility_week[i] = get_volatility(csi, date=i.strftime("%Y-%m-%d"), k=10)
        volatility_month[i] = get_volatility(csi, date=i.strftime("%Y-%m-%d"), k=30)
        volatility_year[i] = get_volatility(csi, date=i.strftime("%Y-%m-%d"), k=365)

    current_date = i
```

```python
def month2date(month):
    (m, d) = month.split(".")
    return datetime.datetime(int(m), int(d), 1)
```

```python
money.index = money["month"].map(lambda x: month2date(x))
cpi.index = cpi["month"].map(lambda x: month2date(x))
```

```python
volatility_df_new = volatility_df.copy()
volatility_df_new["m0_yoy"] = np.zeros(volatility_df_new.shape[0])
volatility_df_new["m1_yoy"] = np.zeros(volatility_df_new.shape[0])
volatility_df_new["m2_yoy"] = np.zeros(volatility_df_new.shape[0])
volatility_df_new["cpi"] = np.zeros(volatility_df_new.shape[0])
```

```python
def last_month(date1, date2):
    """
    如果date2是date1的上一个月，返回True
    """
    if (date1.month == date2.month + 1) and (date1.year == date2.year):
        return True
    elif (date1.month == 1) and (date2.month == 12) and (date1.year == date2.year + 1):
        return True
```

```python
for i in volatility_df_new.index:
    for j in cpi.index:
        if last_month(i, j):
            volatility_df_new.loc[i, "cpi"] = cpi.loc[j, "cpi"]
```

```python
for i in volatility_df_new.index:
    for j in money.index:
        if last_month(i, j):
            volatility_df_new.loc[i, "m0_yoy"] = money.loc[j, "m0_yoy"]
            volatility_df_new.loc[i, "m1_yoy"] = money.loc[j, "m1_yoy"]
            volatility_df_new.loc[i, "m2_yoy"] = money.loc[j, "m2_yoy"]
```

```python
volatility_df_new[["m0_yoy", "m1_yoy", "m2_yoy"]] = volatility_df_new[["m0_yoy", "m1_yoy", "m2_yoy"]].apply(pd.to_numeric)
```

# 词频与波动率之间的关系

```python
with open("./word_count.json", 'r') as f:
    word_cnt = json.load(f)
```

```python
with open("./words_rank.txt", 'r') as f:
    word_select = [line.strip() for line in f.readlines()]
```

```python
word_count = {}
for word in word_select:
    word = word.decode('utf-8')
    word_count[word] = {}
    if word not in word_cnt:
        continue
    for date, value in word_cnt[word].iteritems():
        date = datetime.datetime.strptime(date, "%Y-%m-%d")
        key = "%s.%s"%(date.year, date.month)
        if not word_count[word].has_key(key):
            word_count[word][key] = 0
        word_count[word][key] += value
```

```python
word_df = pd.DataFrame()
for word in word_count:
    temp_df = pd.DataFrame({"date": word_count[word].keys(), word: word_count[word].values()})
    if word_df.empty:
        word_df = temp_df
    else:
        word_df = word_df.merge(temp_df, how='outer')
```

```python
word_df.drop(word_df.index[0], inplace=True)
```

```python
word_df.head()
```

```python
for i in volatility_df_new.index:
    year, month = i.year, i.month
    word_df.loc[word_df["date"] == "%s.%s"%(year, month), "date"] = i
```

```python
word_df.set_index('date', inplace=True)
```

```python
volatility_df_new_word = volatility_df_new.join(word_df)
```
