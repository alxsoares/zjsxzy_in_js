```python
df.sort("Exchange", inplace=True)
```

```python
g = sns.barplot(x="Name", y="Sharpe", data=df, hue="Exchange")
for item in g.get_xticklabels():
    item.set_rotation(90)
    item.set_fontsize(8)
for bar in g.patches:
    bar.set_width(1)
```
