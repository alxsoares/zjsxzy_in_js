```python
def fetch_from_wind(csv_dir, symbols, start_date, end_date):
    cols = ["open", "high", "low", "close", "amt"]
    for s in symbols:
        filename = "%s/%s.csv"%(csv_dir, s)
        print filename
        if os.path.exists(filename):
            continue
        raw_data = w.wsd(s, cols, beginTime=start_date, endTime=end_date)
        dic = {}
        for data, field in zip(raw_data.Data, raw_data.Fields):
            if str.lower(str(field)) == "amt":
                field = "volume"
            dic[str.lower(str(field))] = data
        df = pd.DataFrame(dic)
        df["date"] = pd.to_datetime(raw_data.Times)
        df["date"] = df["date"].map(lambda x: x.strftime('%Y-%m-%d'))
        df = df[["date", "open", "high", "low", "close", "volume"]]
        df.dropna(subset=['close'], inplace=True)
        assert(df.shape[0] != 0)
        df.to_csv(filename, index=False)
```

# 4. 计算指数与成分股月度beta

```python
symbol = "001979.SZ"
```

```python
sym_df = pd.read_csv("%s/stocks/%s.csv"%(DATA_DIR, symbol))
```

```python
sym_df.index = pd.to_datetime(sym_df['date'], format="%Y-%m-%d")
```

```python
sym_df['return'] = sym_df['close'].pct_change()
```

```python
sym_df = sym_df[sym_df.index >= datetime.datetime.strptime(start_date, "%Y-%m-%d")]
```

```python
sym_df.head()
```

```python
sym_df['close'].plot()
```

```python
date = month_ind_var_df.index[0]
stock_return = sym_df['return'][sym_df.index <= date]
index_return = ind_df['return'][ind_df.index <= date]
```

```python
np.cov(stock_return, index_return)[0][1] / month_ind_var_df[0]
```

```python
betas = []
for i, date in enumerate(month_ind_var_df.index):
    stock_return = sym_df.loc[(sym_df.index.year == date.year) & (sym_df.index.month == date.month), 'return']
    index_return = ind_df.loc[(ind_df.index.year == date.year) & (ind_df.index.month == date.month), 'return']
    if stock_return.shape == index_return.shape:
        # 计算beta系数
        beta = np.cov(stock_return, index_return)[0][1] / month_ind_var_df[i]
        betas.append(beta)
    else:
        print symbol, date, stock_return.shape, index_return.shape
        betas.append(np.nan)
```

```python
df[symbol] = betas
```

```python
corr = [np.corrcoef(month_ind_pri_df, 1/df['mean beta'])[0][1]]
for k in range(1, month_ind_pri_df.shape[0] / 2):
    temp = np.corrcoef(month_ind_pri_df[k:], 1/df['mean beta'][:-k])[0][1]
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(month_ind_pri_df[k:], 1/df['mean beta'][:-k])
    print k, temp, r_value**2
    corr.append(temp)
```

```python
df['ratio beta'] = np.zeros(df.shape[0])
for i in df.index:
    high_beta = (df.ix[i] > 1).sum()
    low_beta = (df.ix[i] < 1).sum()
#     print high_beta, low_beta, high_beta * 1.0 / low_beta
    df.loc[i, 'ratio beta'] =  low_beta * 1.0 / high_beta
```

```python
corr = [np.corrcoef(month_ind_pri_df, df['ratio beta'])[0][1]]
for k in range(1, month_ind_pri_df.shape[0] / 2):
    temp = np.corrcoef(month_ind_pri_df[k:], df['ratio beta'][:-k])[0][1]
    print k, temp
    corr.append(temp)
```
